```python
import numpy as np
import scipy.io.wavfile
```

```python
x = np.random.randint(1000) * np.sin(2 * np.random.randint(10**4) * np.pi * t)
t = np.arange(0, 10, 0.0001)
for i in range(1000):
    x += np.random.randint(1000) * np.sin(2 * np.random.randint(100**4) * np.pi * t)
```

```python
import matplotlib.pyplot as plt

plt.plot(t, x)
```

```python
scipy.io.wavfile.write('wav1.wav', 44100, x.astype(np.int16))
```
