```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wf
```

```python
sampling_rate, samples = wf.read('sound.wav')

t = np.arange(0, len(samples)/sampling_rate, 1/sampling_rate)
```
